```python
%%writefile app.py
import streamlit as st
import requests
st.write("APP STARTED")
st.set_page_config(page_title="AQI Predictor")

st.title("ðŸŒ« Air Quality Prediction System")

FEATURES = [
    'PM2.5','PM10','NO','NO2','NOx','NH3',
    'CO','SO2','O3','Benzene','Toluene','Xylene'
]

st.subheader("Enter last 7 days data")

data = []
for day in range(7):
    with st.expander(f"Day {day+1}", expanded=(day == 0)):
        row = []
        for f in FEATURES:
            val = st.number_input(
                f"{f}",
                min_value=0.0,
                step=0.1,
                key=f"{f}_{day}"
            )
            row.append(val)
        data.append(row)

if st.button("Predict AQI"):
    try:
        response = requests.post(
            "http://127.0.0.1:8000/predict",
            json={"data": data},
            timeout=5
        )

        if response.status_code == 200:
            res = response.json()
            st.success(f"Predicted AQI: {res['AQI']}")
            st.info(f"Category: {res['Category']}")
        else:
            st.error("Backend error")

    except Exception as e:
        st.error("FastAPI backend not running")
```
